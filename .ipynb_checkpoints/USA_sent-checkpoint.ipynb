### Defining election data

```python
# initializing dataframe with election data
data_elec = {
    'election_dates': ['2020-11-03','2016-11-08 ', 
                       '2012-11-06', '2008-11-04',
                       '2004-11-02', '2000-11-07',
                       '1996-11-05', '1992-11-03'],
    'change_in_government': [1,1,0,1,0,1,0,0]}

OLS_data = pd.DataFrame(data_elec)
```

### Average sentiment scores (3 month interval)
Computing average sentiment scores for the 3 months leading up to the election date: 

```python
data_sent = pd.read_csv('/Users/astakettel/Desktop/ISDS/GitHub/ISDS-sentiment-analysis/Webscraping/EN_clean_sent.csv')
```

```python
# Ensure the 'date' column in your DataFrame is in datetime format
data_sent['date'] = pd.to_datetime(data_sent['date'])
```

```python
#1. I start by defining the time periods i want to compute the average sentiments of

# List of time periods (start_date, end_date)
time_periods = [
    ('2020-08-03', '2020-11-03'),
    ('2016-08-08', '2016-11-08'),
    ('2012-08-06', '2012-11-06'),
    ('2008-08-04', '2008-11-04'),
    ('2004-08-02', '2004-11-02'),
    ('2000-08-07', '2000-11-07'),
    ('1996-08-05', '1996-11-05'),
    ('1992-08-03', '1992-11-03')]

# 2. Then i define a function to compute the average sentiment scores 
    # found in the articles written within a defined interval: 

def compute_average_sentiment(df, start_date, end_date):
    # Convert string dates to datetime if necessary
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    # Filter DataFrame for the specific time period
    filtered_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]
    
    # Calculate the average sentiment
    average_sentiment = filtered_df['sentiment'].mean()
    
    return average_sentiment

# 3. Then i loop through each interval to get the average sentiment for each interval: 

# Dictionary to store the results for each time period
average_sentiments = {}

for start_date, end_date in time_periods:
    avg_sentiment = compute_average_sentiment(data_sent, start_date, end_date)
    average_sentiments[(start_date, end_date)] = avg_sentiment
```

```python
# Now i store the average sentiment scores in a list: 
sent_3M = list(list(average_sentiments.values()))

# And append the list as a column in the dataframe OLS_data:
OLS_data['sent_3M'] = sent_3M
```

# Example code for OLS regression

```python
# As we dont have articles from all time periods we remove the rows that contain missing information: 
OLS_data_sub = OLS_data.dropna()
```

```python
import pandas as pd
import statsmodels.api as sm
# Define the dependent and independent variables
X = OLS_data_sub[['return_3M','sent_3M']]  # Independent variables (X1, X2)
Y = OLS_data_sub['change_in_government']           # Dependent variable (Y)

# Add a constant to the independent variables (this adds the intercept term)
X = sm.add_constant(X)

# Perform the OLS regression
model = sm.OLS(Y, X).fit()

# Print the OLS regression results
print(model.summary())
```

**Model Summary:**

Dep. Variable: change_in_government
- This is the dependent variable being predicted, which, in this case, represents a change in government (binary outcome).

R-squared: 0.984
- This indicates that 98.4% of the variance in the dependent variable is explained by the independent variables (return_3M and sent_3M). A high R-squared value indicates a strong fit, but with a small sample size (4 observations), this should be interpreted cautiously.

Adj. R-squared: 0.951
- The adjusted R-squared accounts for the number of predictors in the model. It is slightly lower than R-squared, but still high at 95.1%.

No. Observations: 4
- This indicates that the model is based on only 4 data points, which is a very small sample size and can lead to unreliable estimates.

Df Residuals: 1
- Degrees of freedom for residuals are calculated as No. Observations - No. of Parameters. Since there are 2 independent variables (return_3M and sent_3M), and a constant term, we have 4 - 3 = 1 residual degree of freedom.

Df Model: 2
- This corresponds to the number of independent variables (predictors) in the model.

**Coefficients**

Stock Market Impact (return_3M):
- Although the stock market return coefficient is positive, meaning better stock market performance is associated with a higher probability of a government change, the statistical insignificance suggests that more data is needed to confidently assess this relationship.

Sentiment Impact (sent_3M):
- A more positive media sentiment appears to reduce the likelihood of a government change, implying that a favorable media environment may help incumbent governments. However, the statistical insignificance again calls for caution in interpreting this result without further data.



# Logistic regression

```python
import pandas as pd
import statsmodels.api as sm

# Assuming your DataFrame is 'OLS_data_'
# Independent variable: 'sent_3M' (sentiment score)
# Dependent variable: 'change_in_government' (binary: 1 = change, 0 = no change)

# Define independent and dependent variables
X = OLS_data_sub[['return_3M','sent_3M']]  # Independent variables (X1, X2)
Y = OLS_data_sub['change_in_government']  # Government change as the dependent variable

# Add a constant to the independent variable (for the intercept)
X = sm.add_constant(X)

# Fit the Logistic Regression model
logit_model = sm.Logit(Y, X).fit()

# Print the summary of the logistic regression results
print(logit_model.summary())
```
